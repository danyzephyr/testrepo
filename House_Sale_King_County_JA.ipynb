```python
from pyodide.http import pyfetch

async def download(url, filename):
    response = await pyfetch(url)
    if response.status == 200:
        with open(filename, "wb") as f:
            f.write(await response.bytes())
```

### Question 2

Drop the columns "id" and "unnamed: 0" from axis 1 using the method drop(), then use the method describe() to obtain a statistical summary of the data.

```python
#remove requested columns
df.drop(['Unnamed: 0', 'id'], axis=1, inplace=True)

# Let's take a look at the updated DataFrame
df.describe()
```

```python
#we can see we have missing values from bedrooms and bathrooms
print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```

```python
#replace missing values from bedroom with mean of the column using the method replace. don't forget to set the inplace parameter to true
mean=df['bedrooms'].mean()
df['bedrooms'].replace(np.nan,mean, inplace=True)
```

```python
#also replace NaN values for bathroom with mean value, using the method replace. don't forget to set the inplace parameter to true
mean=df['bathrooms'].mean()
df['bathrooms'].replace(np.nan,mean, inplace=True)
```

```python
#there should be no NaN values anymore
print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```
